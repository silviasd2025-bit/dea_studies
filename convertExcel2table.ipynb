```python
import pandas as pd
from datetime import datetime
import os
from google.colab import drive
drive.mount('/content/drive')
file_path = "/content/drive/MyDrive/data/20250914_DEA_EnergySecurityPaper2.xlsx"
file_path = "/content/drive/MyDrive/data/20250930_DEA_EnergySecurityPaper 2_3output_import_export.xlsx"

i_installed_capacity_per_capita = pd.read_excel(file_path,sheet_name='I1-Installed Capacity Capita', nrows=21,  usecols= 'A:Y')
i_energy_diversification        = pd.read_excel(file_path,sheet_name='I2-Energy diversification HHI', nrows=21,  usecols= 'A:Y')
i_energy_import                 = pd.read_excel(file_path,sheet_name='I3-Energy import', nrows=21,  usecols= 'A:Y')
i_perc_renewable                = pd.read_excel(file_path, sheet_name='I4-%Renewable', nrows=21,  usecols= 'A:Y')

o_energy_export                 = pd.read_excel(file_path,sheet_name='O3-Energy export', nrows=21,  usecols= 'A:Y')
o_mwh_capita                    = pd.read_excel(file_path, sheet_name='O1-MWh capita', nrows=21,  usecols= 'A:Y')
o_gdp_capita                    = pd.read_excel(file_path, sheet_name='O2-GDP_capita', nrows=21,  usecols= 'A:Y')
o_carbon_intensity              = pd.read_excel(file_path, sheet_name='O4-carbon intensity', nrows=21,  usecols= 'A:Y')
```

```python
# Rename the first column in each DataFrame to 'dmu' for merging
i_installed_capacity_per_capita.rename(columns={i_installed_capacity_per_capita.columns[0]: 'dmu'}, inplace=True)
#display(i_installed_capacity_per_capita.head())
i_energy_diversification.rename(columns={i_energy_diversification.columns[0]: 'dmu'}, inplace=True)
i_energy_import.rename(columns={i_energy_import.columns[0]: 'dmu'}, inplace=True)
i_perc_renewable.rename(columns={i_perc_renewable.columns[0]: 'dmu'}, inplace=True)

o_mwh_capita.rename(columns={o_mwh_capita.columns[0]: 'dmu'}, inplace=True)
o_gdp_capita.rename(columns={o_gdp_capita.columns[0]: 'dmu'}, inplace=True)
o_energy_export.rename(columns={o_energy_export.columns[0]: 'dmu'}, inplace=True)
o_carbon_intensity.rename(columns={o_carbon_intensity.columns[0]: 'dmu'}, inplace=True)


```

```python
#o_inverse_carbon_intensity = o_carbon_intensity.copy()
#o_inverse_carbon_intensity.dmu = o_carbon_intensity.dmu
#for year in range(2000,2024):
#  o_inverse_carbon_intensity[year] = 1 / o_carbon_intensity[year]
#display(o_inverse_carbon_intensity.head())
#display(o_carbon_intensity.head())
```

```python
# Melt the dataframes to long format
i_installed_capacity_per_capita_melted = i_installed_capacity_per_capita.melt(id_vars='dmu', var_name='year', value_name='installed_capacity_per_capita')
#display(i_installed_capacity_per_capita_melted.head(25))
#print(i_installed_capacity_per_capita_melted.shape)
i_energy_diversification_melted = i_energy_diversification.melt(id_vars='dmu', var_name='year', value_name='energy_diversification')
i_energy_import_melted = i_energy_import.melt(id_vars='dmu', var_name='year', value_name='energy_import')

i_perc_renewable_melted = i_perc_renewable.melt(id_vars='dmu', var_name='year', value_name='percentage_renewable')

o_mwh_capita_melted = o_mwh_capita.melt(id_vars='dmu', var_name='year', value_name='mwh_capita')
o_gdp_capita_melted = o_gdp_capita.melt(id_vars='dmu', var_name='year', value_name='gdp_capita')
o_energy_export_melted = o_energy_export.melt(id_vars='dmu', var_name='year', value_name='energy_export')
o_carbon_intensity_melted = o_carbon_intensity.melt(id_vars='dmu', var_name='year', value_name='carbon_intensity')
#o_inverse_carbon_intensity_melted = o_inverse_carbon_intensity.melt(id_vars='dmu', var_name='year', value_name='inverse_carbon_intensity')
```

```python
merged_df = pd.merge(i_installed_capacity_per_capita_melted, i_energy_diversification_melted, on=['dmu', 'year'], how='outer')
#display(merged_df.head())
#print(merged_df.shape)
```

```python
#display(i_energy_export_melted.head())
#print(i_energy_export_melted.shape)
merged_df = pd.merge(merged_df, i_energy_import_melted, on=['dmu', 'year'], how='outer')
#display(merged_df.head())
#print(merged_df.shape)
```

```python
# Merge the dataframes


merged_df = pd.merge(merged_df, i_perc_renewable_melted, on=['dmu', 'year'], how='outer')

merged_df = pd.merge(merged_df, o_mwh_capita_melted, on=['dmu', 'year'], how='outer')
merged_df = pd.merge(merged_df, o_gdp_capita_melted, on=['dmu', 'year'], how='outer')
merged_df = pd.merge(merged_df, o_energy_export_melted, on=['dmu', 'year'], how='outer')
merged_df = pd.merge(merged_df, o_carbon_intensity_melted, on=['dmu', 'year'], how='outer')
#merged_df = pd.merge(merged_df, o_inverse_carbon_intensity_melted, on=['dmu', 'year'], how='outer')

# Select the desired columns
final_df = merged_df[['dmu', 'year', 'installed_capacity_per_capita',
                     'energy_diversification', 'energy_import', 'percentage_renewable',
                     'mwh_capita', 'gdp_capita', 'energy_export', 'carbon_intensity']]

# Display the resulting DataFrame
display(final_df.head())
#df2 = final_df.iloc[:100]
#print(df2)
```
